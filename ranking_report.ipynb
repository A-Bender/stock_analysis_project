```python
# Function to extract earnings

def fetch_earnings(symbol):
    url = f'{BASE_URL}?function=EARNINGS&symbol={symbol}&apikey={API_KEY}&entitlement=delayed'
    response = requests.get(url).json()
    return response


# Function to extract financial statement

def fetch_financial_statement(symbol):
    url = f'{BASE_URL}?function=INCOME_STATEMENT&symbol={symbol}&apikey={API_KEY}&entitlement=delayed'
    response = requests.get(url).json()
    return response

# Function to extract balance sheet

def fetch_balance_sheet(symbol):
    url = f'{BASE_URL}?function=BALANCE_SHEET&symbol={symbol}&apikey={API_KEY}'
    response = requests.get(url).json()
    return response

# Function to extract current stock price

def fetch_current_price(symbol):
    url = f'{BASE_URL}?function=GLOBAL_QUOTE&symbol={symbol}&apikey={API_KEY}&entitlement=delayed'
    response = requests.get(url).json()
    return response

    try:
        price = float(response['Global Quote']['05. price'])
        return price
    except (KeyError, ValueError):
        return None
```
